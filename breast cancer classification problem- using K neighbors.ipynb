```python
import numpy as np
import pandas as pd
from sklearn import preprocessing, model_selection, neighbors
```

```python
df=pd.read_csv("C:\Users\User\Desktop\\breast-cancer-wisconsin.txt")
```

```python
df.replace("?",-99999,inplace=True)  # the dataframe contains "?" , so replace it by -9999
```

```python
df.drop(["id"], axis=1, inplace=True) # dropping the id column as it is not neccessary 
```

```python
df
```

```python
x=np.array(df.drop(["class"],axis=1))
y=np.array(df["class"])

```

```python
x_train,x_test,y_train,y_test= model_selection.train_test_split(x,y,test_size=0.20,random_state=41)
```

```python
clf=neighbors.KNeighborsClassifier()
clf.fit(x_train,y_train)
accuracy=clf.score(x_test,y_test)
print accuracy
```
